```python
import pandas as pd
import numpy as np
import math

train_features = pd.read_csv('test_features.csv')
train_labels = pd.read_csv('sample.csv')

labels = ['EtCO2', 'PTT', 'BUN', 'Lactate', 'Temp',
       'Hgb', 'HCO3', 'BaseExcess', 'RRate', 'Fibrinogen', 'Phosphate',
       'WBC', 'Creatinine', 'PaCO2', 'AST', 'FiO2', 'Platelets', 'SaO2',
       'Glucose', 'ABPm', 'Magnesium', 'Potassium', 'ABPd', 'Calcium',
       'Alkalinephos', 'SpO2', 'Bilirubin_direct', 'Chloride', 'Hct',
       'Heartrate', 'Bilirubin_total', 'TroponinI', 'ABPs', 'pH']

_labels = ['pid', 'Age', 'EtCO2_min', 'EtCO2_max', 'PTT_min', 
           'PTT_max', 'BUN_min', 'BUN_max', 'Lactate_min', 'Lactate_max', 
           'Temp_min', 'Temp_max', 'Hgb_min', 'Hgb_max', 'HCO3_min', 'HCO3_max', 
           'BaseExcess_min', 'BaseExcess_max', 'RRate_min', 'RRate_max', 
           'Fibrinogen_min', 'Fibrinogen_max', 'Phosphate_min', 'Phosphate_max', 
           'WBC_min', 'WBC_max', 'Creatinine_min', 'Creatinine_max', 'PaCO2_min', 
           'PaCO2_max', 'AST_min', 'AST_max', 'FiO2_min', 'FiO2_max', 
           'Platelets_min', 'Platelets_max', 'SaO2_min', 'SaO2_max', 
           'Glucose_min', 'Glucose_max', 'ABPm_min', 'ABPm_max', 
           'Magnesium_min', 'Magnesium_max', 'Potassium_min', 'Potassium_max', 
           'ABPd_min', 'ABPd_max', 'Calcium_min', 'Calcium_max', 
           'Alkalinephos_min', 'Alkalinephos_max', 'SpO2_min', 
           'SpO2_max', 'Bilirubin_direct_min', 'Bilirubin_direct_max', 
           'Chloride_min', 'Chloride_max', 'Hct_min', 'Hct_max', 
           'Heartrate_min', 'Heartrate_max', 'Bilirubin_total_min', 
           'Bilirubin_total_max', 'TroponinI_min', 'TroponinI_max', 
           'ABPs_min', 'ABPs_max', 'pH_min', 'pH_max']


#print(train_features[train_features['pid'].isin([2])])

def calculate_min_max_value(mylist):
    newlist = []
    for item in mylist:
        if not math.isnan(item):
            newlist.append(item)
    if len(newlist) == 0:
        return np.nan, np.nan
    else:
        return sorted(newlist)[0], sorted(newlist)[-1]
    
def reduce_frame(myframe):
    df = []
    df.append(myframe['pid'].values[0])
    df.append(myframe['Age'].values[0])
    for item in labels:
        (minv, maxv) = calculate_min_max_value(myframe[item].values)
        df.append(minv)
        df.append(maxv)
    return df

def reduce_tot_frame(myframe, pid_list):
    data=open('test_features_r.csv','w')
    out_to_file(_labels, data)
    
    for item in pid_list:
        mylist = reduce_frame(myframe[myframe['pid'].isin([item])])
        out_to_file(mylist, data)
        print(item)
    data.close()   
    
        
def out_to_file(mylist, myfile):
    for i in range(0, len(mylist)):
        if i != len(mylist)-1:
            print(mylist[i],end='',file=myfile)
            print(',',end='',file=myfile)
        else:
            print(mylist[i],end='\n',file=myfile)
    

pid_list = train_labels['pid'].values
reduce_tot_frame(train_features, pid_list)
```
